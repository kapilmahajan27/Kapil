Tripadvisor scraping     

The objective of this assignment is to scrape Tripadvisor page for hotels in a city and create a dataframe containing various details about the hotels.

      



### Importing required packages

```python
import requests
from requests import get
from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib.request
import pandas as pd
import re
import requests
```

### Defining the URL for hotel search in Jodhpur city and reading in the HTML for that page using BeautifulSoup

```python
url = 'https://www.tripadvisor.in/Hotels-g297668-Jodhpur_Jodhpur_District_Rajasthan-Hotels.html'
url2 = requests.get(url) 
soup = BeautifulSoup(url2.text,'html.parser')
```

### Looping through the URLs for each hotel one by one and extracting the details - name,link,price and savings

```python
hotels_all = []
for i in soup.find_all('div',attrs = {'class' : 'listing'}):
    hotels = {}
    hotels['name'] = i.find('div',attrs = {'class' : 'listing_title'}).text
    try:
        hotels['price'] = i.find('div',attrs = {'class' : 'price-wrap'}).div.text
    except:
        hotels['price'] = 'None'
    try:
        hotels['link'] = 'https://www.tripadvisor.in' + i.find('div',attrs = {'class' : 'listing_title'}).a['href'] 
    except:
        hotels['link'] = 'None'
    try:
        hotels['savings'] = i.find('div',attrs = {'class' : 'xwrap ui_columns is-gapless'}).div.text
    except:
        hotels['savings'] = 'None'
    hotels_all.append(hotels)
```

### Creating a dataframe with the details

```python
df = pd.DataFrame(hotels_all)
df.head()
```

### Looping through URLs for each hotel and extracting address details, ratings, review counts and amenities

```python
addr1 = []
for x in df['link']:
    url = requests.get(x)
    soup = BeautifulSoup(url.text,'html.parser')
    addr = {}
    addr['link'] = x
    try:
        addr['review_counts'] = soup.find('span',attrs = {'class':'reviewCount'}).text
    except:
        addr['review_counts'] = 'None'
    try:
        addr['rating'] = soup.find('span',attrs = {'class' : 'overallRating'}).text
    except:
        addr['rating'] = 'None' 
    try:
        addr['address'] = soup.find('span',attrs = {'class' : 'detail'}).text
    except:
        addr['address'] = 'None'    
    ls = []    
    for x in soup.find_all('div',attrs = {'class':'highlightedAmenity detailListItem'}):
        ls.append(x.text)
        addr['amenities'] = ls   
    addr1.append(addr)
```

### Creating a dataframe

```python
address_data = pd.DataFrame(addr1)
address_data.head()
```

### Joining the two dataframes on link

```python
final1 = pd.merge(df,address_data, how = 'left',left_on = 'link',right_on = 'link')
final1
```

### Reordering the columns

```python
final = final1[['name','address','link','price','rating','review_counts','savings','amenities']]
final.head()
```

### Some cleanups

```python
final['savings'] = final['savings'].str.replace('SAVE ','')
final
```
